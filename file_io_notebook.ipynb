```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("employees_processed.csv")
print("Processed DataFrame: ")
print(df)
```

```python
#Visualize the normalized salary by department
plt.figure(figsize=(8, 5))
plt.scatter(df["department"], df["salary_normalized"], color="blue", label="Normalized Salary")
plt.xlabel("Department")
plt.ylabel("Normalized Salary")
plt.title("Normalized Salary by Department")
plt.legend()
plt.show()
```
